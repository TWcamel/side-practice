```python
#----------------------py填充概念------------------------#

#基本資料類型：布林(bool)、整數(integer)、浮點數(float)與字串(string)
#大多數時候我們需要將這些物件分割與結合成特定格式，他們就是我們的鋼鋸與膠槍

#1. #序列結構
#除字串外，python有另外兩種序列結構： tuple, 串列(list)
#與字串不同的是，他們可以用不同類型的元素，當中的每個元素可以是任何python的物件

#1.1 #為何python同時擁有兩種序列結構：tuple與list?
#Tuple是不可變的，當指派一個元素給Tuple序列結構後，就不可更改了。
#list是可變的，可以盡情的插入與刪除list序列結構當中的元素


#----------------------py填充概念結束------------------------#
```

```python
#----------------------串列練習------------------------#

##用list()將其他的資料類型轉成串列序列結構(list)
a_tuple = ('ready', 'fire', 'aim')
print(list(a_tuple),'\n','\n','----')

##練習於ch2所學到的'用split()來分割字串'
birthday = '1/6/2019'
birthday = birthday.split('/')
print(birthday,'\n','\n','----')

##串列的串列
small_birds = ['humingbird', 'finch']
big_birds = [3, 'passenger pigon', 'Norwegain Blue']
all_birds = [small_birds, big_birds, 'macaw', 777]
print(all_birds)
print(all_birds[0], ', ', all_birds[1][-1] ,'\n','\n','----') #索引值都於位移值上操作!

##用位移值來更改一個項目
all_birds[1][0] = 9
all_birds = all_birds
print(all_birds ,'\n','\n','----')

##使用append()將元素附加到結尾
all_birds.append('A new bird!')
print(all_birds ,'\n','\n','----')

##使用insert(), del,remove() 來更改list序列結構中的元素(項目)
#insert()
extinct_birds = ['Harpo', 'Groucho']
all_birds.insert(0, extinct_birds)
print(all_birds)
#del
del all_birds[2]
print(all_birds)
#remove()
#all_birds.remove('Groucho') #會報錯
all_birds.remove('macaw')
print(all_birds,'\n','\n','----')

##用index()尋找某元素的位移值
all_birds.index('A new bird!')

##用join()轉換序列結構：list to sting 
marxes = ['cheeseburger', 'Chice', 'Harpo']
print(type(marxes))
marxes = ', ' .join(marxes)
print(type(marxes), marxes,'\n','\n','----')

```

```python
##運用ch2學到的便利貼與物件方法來指定序列結構
a = [1, 2, 3]
print(a)
b = a
print(b)
a[0] = 'surprise!!'
print('a:', a)

#此時b與a相同!
print('b:', b)

#稍作修改
b[0] = 'I hate surprise!'
print('revised b[0]', '= I hate surprise!', 'and a is ', a)

#----------------------串列練習結束------------------------#
```

```python
##Tuple (可以想成唯讀檔)
#tuple 與 list很像，都為一系列的任意項目；但與之不同的是，tuple一旦定義後就不可變了。
#用()來建立Tuple

empty_tuple = ()
print(empty_tuple, ',', 'type:', type(empty_tuple),'\n','\n','----')

one_marx = 'Samuel'
print(one_marx, ',', 'type:', type(one_marx),'\n','\n','----')

#python 在印出tuple時，會加上括號，括號只是讓tuple序列結構裡面的元素看得更清楚（可以不必理會）
#因為真正定義tuple的是 , (逗號)
marx_tuple = 'Samuel', 'Chico', 'Harpo'
print(marx_tuple, ',', 'type:', type(marx_tuple),'\n','\n','----')

#tuple可以被一次指派多個變數 (此時可稱為tuple開箱(unpacking))
a,b,c = marx_tuple
print(a, ',', b, ',', 'type:', type(a), type(b))

#----------------------Tuple練習結束------------------------#
```

```python
##字典(dictionary) 以{key: value}雙值序列來表示，雖然與list序列結構很像，但dict不在乎順序，故較像無序結構。
#用{}來建立dict
bierce = {
    "day": "A period of twenty-four hours, mostly misspent",
    "posotive": "Mistaken at the top of one's voice",
    "misfortune": "The kind of fortune that never misses"
}
print(bierce,'\n','\n','----')

#用dict()函式來轉換
#記得dict裡面，key的順序都是任意排列的!
lol = [['a', 'b'], ['c', 'd'], ['e', 'f']]
print(dict(lol), ',', 'type:', type(lol),'\n','\n','----')

#一個含有雙項目tuple的串列(list)
lot = (['a', 'b'], ['c', 'd'], ['e', 'f'])
print(lot, ',', 'type:', type(lot),'\n','\n','----')

#一個含有雙字元sring(字元)的list(串列)
los = ['ab', 'cd', 'ef']
print(los, ',', 'type:', type(los),'\n','\n','----')

#一個含有雙字元字串(string)的tuple
tos = ('ab', 'dc', 'ef')
print(tos, ',', 'type:', type(tos),'\n','\n','----')


```

```python
#利用update()來合併dict
bibles = {
    "day": "A period of twenty-four hours, mostly misspent",
    "posotive": "Mistaken at the top of one's voice",
    "misfortune": "The kind of fortune that never misses"
}
others = {
    'Marx': 'Gruoucho', 'Howard': 'Moe'
}

bibles.update(others)
print(bibles,'\n','\n','----')

#若有相同的字時
same_word = {
    'Howard': 'Moe'
}
others.update(same_word)
print(others,'\n','\n','----')

#用keys(),values()來取得所有的鍵(key)與值(value)
print(bibles.keys())
print(bibles.values(),'\n','\n','----')

#用items()來取得所有的('key'/'value')對
print(others.items(),'\n','\n','----')

#----------------------dict()練習結束------------------------#
```

```python
##集合(set)
#集合就像被移除value，只留下key的dict()一樣，每個key都必須是獨一無二的。
#適用於想知道某東西是否存在
#若想知道某些資訊指派給key最為他的value，則使用dict()

#以set來建立
empty_set = set()
print(empty_set, ',', 'type:', type(empty_set),'\n','\n','----')

#與dict()一樣，集合是無序的!
even_numbers = {0, 2, 4, 6, 8}
print(even_numbers, ',', 'type:', type(even_numbers),'\n','\n','----')

############################小觀念############################

#事實上，[]會建立一個空串列(list)
#但{}不會建立空集合，而是建立一個空的dict()
#這就是為何interpreter將空集合印成set()，而不是{}

############################小觀念############################

##用set轉換其他類型
#此例中，集合裡面只有一個e與一個t(交集)
print(set('letters'), ',', 'type:', type(set('letters')),'\n','\n','----')

##使用in來測試value
#這是set最常用到的地方
drinks = {
    'martini': {'vodka', 'vermouth'},
    'black russian': {'vodka', 'kahlua'},
    'white russian': {'cream', 'kahlua', 'vodka'}
}
print(drinks.items(),'\n','\n','----')
for names, contents in drinks.items():
    if 'vodka' in contents:
        print(names)
print('\n','\n','----')

#如果想喝vodka，但不要cream與vermouth
# &(集合交集運算子)運算子會產生一個set，裡面存有dict()內都有的項目
for name, contents, in drinks.items():
    if 'vodka' in contents and not contents & {'cream', 'vermouth'}:
        print(name)

#----------------------set練習結束------------------------#
```

```python
##結合與運算子
aa = {1, 2}
bb = {2, 3}
print(aa & bb,'\n','\n','----')

#運用intersection()來取得交集
bb = {3, 4}
print(aa.intersection(bb),'\n','\n','----') #空集合以set()來表示

#　聯集
aa = {3, 2}
bb = {2, 4}
print(aa|bb,'\n','\n','----')


#  差集(屬於第一個set，但不屬於第二個set的元素)
print(aa - bb,'\n','\n','----')

#  互斥 (或使用 ^ 或使用symmetric_difference())
print(aa ^ bb,'\n','\n','----')

#  使用 <= 或 issubset() 來檢查一個set是否為另一個set的子集合
print(aa <= aa,'\n','\n','----')

#  使用 => 或 issuperset() 來檢查一個set是否為另一個set的超集合(子集合的相反)
print(aa.issuperset(aa),'\n','\n','----')

#----------------------集合與運算子練習結束------------------------#
```
